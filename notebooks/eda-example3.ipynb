```python
import pandas as pd
import matplotlib.pyplot as plt # lib para plotagem de grficos
import seaborn as sns # lib para plotagem de grficos mais avanados e estilizados

"""
Anlise Exploratria de Dados (EDA  Exploratory Data Analysis) utilizando grficos de heatmap, disperso e boxplot. 
"""

dados = pd.read_csv('resources/veiculos.csv')

# plotando grfico de heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(dados.corr(), annot=True, cmap='coolwarm')
plt.show()

# plotando grfico de disperso -> relao entre km_rodados e preco. o quanto maior o km_rodados, menor o preco
plt.figure(figsize=(10, 6))
sns.scatterplot(x='km_rodados', y='preco', data=dados)
plt.show()

# plotando grfico de boxplot -> relao entre preco e km_rodados -> quanto maior o km_rodados, menor o preco
plt.figure(figsize=(10, 6))
sns.boxplot(x='km_rodados', y='preco', data=dados)
plt.show()
```
